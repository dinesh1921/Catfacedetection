```python
image=cv2.imread('Cat.jpg')
```

```python
plt.imshow(image)
plt.show()
```

```python
cascade = cv2.CascadeClassifier(r'C:\Users\DELL\Downloads\haarcascade_frontalcatface.xml')
```

```python
model=cascade.detectMultiScale(image)
print('Number of detected cat faces:', len(model))
```

```python
for (x,y,w,h) in model:
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(image)
plt.show()
cv2.destroyAllWindows() 
```
